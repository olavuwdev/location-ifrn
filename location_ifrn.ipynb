```python
import folium
from folium import plugins
```

```python
ifrn_nomes = ["IFRN - MOSSORO ", "IFRN - APODI "]
coordenadas_ifrn = [[-5.1868041951011055, -37.314021802067366],[-5.592015332605576, -37.811768027491986]]

ifrn_mapa = folium.Map(location=[ -5.4766,  -37.6282], zoom_start= 10, control_scale=True, tiles = "cartodbpositron")
# folium.TileLayer("cartodbpositron").add_to(ifrn_mapa)
for i in range(len(coordenadas_ifrn)):
    coordenada = coordenadas_ifrn[i]
    nomes = ifrn_nomes[i]
    folium.Marker(coordenada, popup=nomes, icon = folium.Icon(icon = "glyphicon glyphicon-calendar", color = "black", icon_color ="white", prefix = "glyphicon")).add_to(ifrn_mapa)

ifrn_mapa.add_child(folium.LatLngPopup())
ifrn_mapa.save("localizações_ifrn.html")
ifrn_mapa
```
